```python
impressionism = get_image_files('/content/gdrive/MyDrive/Wikiart-dataset/wikiart/Impressionism')
```

```python
impressionism = impressionism[:50]
```

```python
def label_func(f):
  print(f)
  return 'Impressionism' in str(f.parents[0])
```

```python
dls = ImageDataLoaders.from_name_func('/content/gdrive/MyDrive/Wikiart-dataset/wikiart/Impressionism', impressionism, label_func, item_tfms=Resize(300, 300), num_workers=0)
```

```python
 learn = cnn_learner(dls, resnet34, metrics=error_rate)
```

```python
learn.fit_one_cycle(4)
```

```python
 learn.fine_tune(1)
```

```python
    learn.show_results()
```

```python
learn.predict('/content/gdrive/MyDrive/Colab Notebooks/fastai/test_1.jpg')
```
