```python
import cv2
import os
import random
```

```python
root = '/content/gdrive/MyDrive/project-art/datasets/temp/train_2'
```

```python
image_paths_labels = []

for image_name in os.listdir(root):
  image_path = os.path.join(root, image_name)

  label = 'Impressionism' if 'impressionism' in image_name else 'Surrealism' 
  image_paths_labels.append((image_path, label))
```

```python
surrealism = [(path, 0) for (path, label) in image_paths_labels if label == 'Surrealism'][:50]
impressionism = [(path, 1) for (path, label) in image_paths_labels if label == 'Impressionism'][:50]
```

```python
print(len(surrealism))
print(len(impressionism))
```

```python
import random

image_paths_labels = surrealism + impressionism
random.seed(42)
random.shuffle(image_paths_labels)
```

```python
print(image_paths_labels[:10])
```

```python
import numpy as np

data = []
target = []


for image_path, label in image_paths_labels:
  image = cv2.imread(image_path)
  image = cv2.resize(image, (16, 16)).flatten()
  image = np.array(image, dtype='float32')
  data.append(image)
  target.append(target)
```

```python
print(len(data))
print(len(target))
```

```python
# from sklearn.preprocessing import LabelBinarizer
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(data, target, test_size=0.3)
```

```python
print(len(x_train))
print(len(x_test))
print(x_train[0].shape)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=2) # ovr eller multinomial (auto beror på hur många kategorier)
```

```python
print(x_train[0])
print(type(x_train[0]))
print(y_train[0])
print(type(y_train[0]))
```

```python
model.fit(x_train, y_train)
```

```python
model.score()
```
