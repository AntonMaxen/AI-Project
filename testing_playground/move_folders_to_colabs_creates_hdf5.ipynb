```python
categories = {
    'Post_Impressionism': 1,
    'Color_Field_Painting': 2, 
    'Fauvism': 3,
    }
```

```python
images = []
labels = []

start = time.time()
os.chdir('/content/resize_test')

for folder in categories.keys():
  for name in os.listdir(os.path.join(root, folder)):
    image_path = os.path.join(root, folder, name)
    print(image_path)
    image = cv2.imread(image_path)
    image = cv2.resize(image, (256, 256))
    image = img_to_array(image)
    images.append(image)
    labels.append(categories[folder])
    # cv2.imwrite("256.jpg", image)

end = time.time()
print(f"It took : {datetime.timedelta(seconds=int(end - start))}")
```
